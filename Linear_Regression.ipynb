```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
np.set_printoptions(precision=2)
```

```python
data = pd.read_csv('final_cars_datasets.csv', error_bad_lines=False)
```

```python
data.head()
```

**Сведем эту таблицу к данным, имеющим одинаковую марку, модель, год, трансмиссию и т.д. Различия в этих данных будут заключаться только в пробеге и цене.**


```python
hd=data[(data["mark"] == "nissan") & (data["transmission"] == "at") & (data["model"] == "note")& (data["year"] == 2011) & (data["drive"]=="2wd")]
```

```python
hd
```

```python
x=np.array(hd["mileage"]).reshape((-1,1))
y=np.array(hd["price"])
```

```python
y
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

**Коэффициент детерминации**

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```

*Малое значение коэффициента детерминации связано с тем, что входная переменная(пробег) плохо коррелирует с выходной переменной(цена). Так, можно увидеть, что в некоторых случаях при большом значении пробега, цена куда больше, чем при значении пробега поменьше.*

```python
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
```
